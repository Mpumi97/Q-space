```python
x = [29, 9, 10, 38, 16, 26, 50, 10, 30, 33, 43, 2, 39, 15, 44, 29, 41, 15, 24,
50]
y = [65, 7, 8, 76, 23, 56, 100, 3, 74, 48, 73, 0, 62, 37, 74, 40, 90, 42, 58,
100]
```

```python
x = np.array(x)
y = np.array(y)
```

```python
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_sq_x = np.sum(x**2)
sum_sq_y = np.sum(y**2)
sum_xy = np.sum(x*y)
```

```python
print(f'sum_x = {sum_x}')
print(f'sum_y = {sum_y}')
print(f'sum_xy = {sum_xy}')
print(f'sum_sq_x = {sum_sq_x}')
print(f'sum_sq_y = {sum_sq_y}')
```

```python
n= x.size
```

```python
cov_xy = n * sum_xy -sum_x*sum_y
```

```python
sd_squared = np.sqrt(n*sum_sq_x-(sum_x)**2) * np.sqrt(n*sum_sq_y-(sum_y)**2)
```

## Hello

```python
sd_squared
```

```python
cov_xy
```

```python
r_squared = (cov_xy/sd_squared)**2
```
