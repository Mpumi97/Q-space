```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Matplotlib

Matplotlib is the default plotting and graphing library for Python.

### 1. X-Y Line Plots

```python
x1 = np.arange(0,10.1,0.1)
y1 = x1**2*np.sin(x1)
x2 = np.arange(0,10.1,0.1)
y2 = x2**2*np.cos(x2)
plt.figure(figsize = (5.5,5.5))
plt.plot(x1,y1, 'y.')
plt.plot(x2,y2,'g-.')
plt.savefig('Two graphs.jpg')
plt.tight_layout()
plt.show()
```

### Graph titles, Axis labels, Legends and Formatting

Adding a title to a `matplotlib` graph is done with the title function, after having the used plot function generate your plot before `plt.show()`

```python
x3  = np.arange(0,10.1,0.1)
y3 = x3**3*np.sin(x3)
plt.plot(x3,y3, 'r-')
plt.xlabel('x3')
plt.ylabel('y3')
plt.title('My first graph')
plt.xticks(np.arange(0,10,2))
plt.yticks(np.arange(-400,400, 100))
plt.grid(linestyle='dashed', color = 'k', which = 'major', axis = 'y')
plt.savefig('mygraph.jpg')
plt.show()
```

```python
x1 = np.arange(0,10.1,0.1)
x2 = np.arange(0,10.1,0.1)
x3 = np.arange(0,10.1,0.1)

y1 = x1**2*np.sin(x1)
y2 = x2**2*np.cos(x1)
y3 = x3**2

plt.plot(x1,y1, 'r')
plt.plot(x2,y2, 'g')
plt.plot(x3,y3, 'b')
plt.legend(['1st series', '2nd series', '3rd series'], bbox_to_anchor = (1.05,1))
plt.show()
```

### Multiple Figures and Subplots

```python
plt.plot(x1,y1,'r.')
plt.title('First plot')
plt.figure(2)
plt.plot(x2,y2,'g*')
plt.title('Second plot')
plt.figure(1)
plt.xlabel('Time (s)')
plt.savefig('Mutiple plots.jpg')
plt.show()
```

```python
xa=np.arange(0,10.1,0.1)
ya=xa**2*np.sin(xa)
xb=np.arange(0,10.1,0.1)
yb=xb**2*np.cos(xb)
xc=np.arange(0,10.1,0.1)/10
yc=xc**2*np.sin(xc)
xd=np.arange(10,20.1,0.1)
yd=xd**2*np.sin(xd)
```

```python
plt.subplot(2,2,1)
plt.plot(xa,ya,'g*')
plt.subplot(2,2,2)
plt.plot(xb,yb,'r.')
plt.subplot(2,2,3)
plt.plot(xc,yc,'g--')
plt.subplot(2,2,4)
plt.plot(xd,yd,'b-.')
plt.show()
```

```python
x = np.arange(1,6)
y1 = np.array([5,3,8,7,1])
y2 = np.array([2,3,6,3,5])
```

```python
plt.bar(x,y1, 0.4, align = 'edge', color = 'r')
plt.bar(x,y2, -0.4, align = 'edge', color = 'g')
plt.legend(['First yield', 'Second yield'], bbox_to_anchor = (1.05,1))
plt.title('The Yields', fontsize = 14, fontweight = 'bold')
plt.xlabel('Term', fontsize = 14, fontweight = 'bold')
plt.ylabel('Total yield', fontsize = 14, fontweight = 'bold')
plt.tight_layout()
plt.savefig('Yields.jpg')
plt.show()  
```
