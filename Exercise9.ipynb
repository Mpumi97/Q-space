```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. read data from csv to numpy array
data = np.genfromtxt('data.csv', delimiter = ',', skip_header = 1)

#extract x values (independent variables)
x = data[:,1]

#extract y values (dependent variables)
y = data[:,2]

# 2. Perform a linear regression and output the slope and intercept coefficients for

#evaluating the slope and intercept values
model = np.polyfit(x,y,1)
print(f'The slope and intercept coefficients are {model}')

#obtaining the model 
p_model = np.poly1d(model)
print(p_model)

# 3. Predicting the correspoding y-value for x = 3.25 and
x_pred = 3.25
y_pred = p_model(3.25)

print(f'The y-value corresponding to x = {x_pred} is {round(y_pred, 3)}')

# 4. Calculating the value of r2

n = x.size

#evaluating the sum
sum_x = np.sum(x)
sum_x_2 = np.sum(np.square(x))
sum_y = np.sum(y)
sum_y_2 = np.sum(np.square(y))
sum_xy = np.sum(x*y)

#calculating the r2 value
top = n*sum_xy - sum_x*sum_y
bot_a = np.sqrt(n*sum_x_2 - np.square(sum_x))
bot_b = np.sqrt(n*sum_y_2 - np.square(sum_y))
bot = bot_a*bot_b
R_2 = np.square(top/bot)
print(f"The value of r_2 = {R_2}")


# 5&6. Scatter plot of x vs y values and regression line

regline = np.linspace(-6,8,100)
y_reg = p_model(regline)

plt.scatter(x,y, color = 'blue')
plt.plot(regline, y_reg, color = 'green')
plt.xlabel('x values', fontsize =14, fontweight = 'bold')
plt.ylabel('y values', fontsize = 14, fontweight = 'bold')
plt.title('Scatter plot of x vs y values', fontsize = 14, fontweight = 'bold')
plt.savefig('Scatter plot.jpg')
plt.show()
```
